### Goals and Analysis
The overal goal with the project is to be able to predict a countries happiness by using other statistical information representing the country. If given a countries level of social support or GDP per captia, I want to be able to predict the happiness of the country.

What is the happiness scores? 
The happiness score is data collects from the Gallup World Poll$^2$. The scores are calculated from a survey containing "main life evaluation questions". The survey asks the respondent to answer 10 if life is perfect and 0 meaning the worst possible situation. The happiness score is the total of all the factors summed together. There is also a Dystopia factor that is an imaginary country used as a benchline against each country's self. This factor is calculated by taking the values equal to the lowest national average for each of the six factors. This column has no impact to the actual score, so for the sake of this project I will be ignoring this factor. Lastly, the 2018 data has a few less factors than the earlier data sets, but they contain the factors that I wish to focus on in this project so I will be using the 2018 data set and leaving the 15, 16, and 17 data in as reference to show the other possible factors that could've been used.

My prediction of which factor that is going to be the most impactful on a countries happiness is the Social Support statistic. Just from my personal belief, if a country supports it's people with well funded public schools, hospitals, librarys, transportation, and other public amenities, it would relieve a lot of unneeded stress on the lower and middle class citizens. Instead of worrying where your children can go to school or what will happen if you get sick, you can instead focus on things that make you happy.

### Finding outliers 
Now that we have a clean data set with easy to use names and no null values, I can look for any outliers in the data by using boxplots and seeing how many points lay outside the whiskers (if any).

```python
# Box plot all the columns besised rank and names
figure, axs = plt.subplots(4, 2, figsize=(12,8))

axs[0][0].boxplot(x = df["Score"])
axs[0][0].set_title("Score")

axs[0][1].boxplot(x = df["GDP"])
axs[0][1].set_title("GDP")

axs[1][0].boxplot(x = df["Social"])
axs[1][0].set_title("Social")

axs[1][1].boxplot(x = df["Life_Expectancy"])
axs[1][1].set_title("Life_Expectancy")

axs[2][0].boxplot(x = df["Freedom"])
axs[2][0].set_title("Freedom")

axs[2][1].boxplot(x = df["Generosity"])
axs[2][1].set_title("Generosity")

axs[3][0].boxplot(x = df["Corruption"])
axs[3][0].set_title("Corruption")

figure.set_figheight(20)
figure.set_figwidth(10)
plt.show();
```

There are a few outliers in Social and Generoisty, but the most by an extreme amount is in the Corruption column. To fix this I will find the interquartile range and use that to determine which data points are the outliers. Once I find these data points, I will replace them with the average from the entire corruption column.

IQR Reference: https://medium.com/@MaheshGadakari/understanding-outliers-for-beginners-79f1c649c20f

```python
# Interquartile Range
Q1 = df["Corruption"].quantile(0.25)
Q3 = df["Corruption"].quantile(0.75)
IQR = Q3 - Q1

# Get the bounds
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

# Now I can find all the data outside the lower and upper ranges
outside =  df["Corruption"] < lower
outside +=  df["Corruption"] > upper

outliers = df[outside]
outliers
```

```python
# Now swap all the values for the Corruption average
corrupt_average = df["Corruption"].mean()

for i in outliers.index:
    df.at[i,'Corruption'] = corrupt_average

df[outside]
```

## Now we have clean data and can begin
First I am going to create a simple function that I will call with each column data set. This function will plot a line plot and a historgram plot with both the regular data and the normalized data for both columns submitted.

```python
def plot_my_graph(data1, data2, title):
    """  
    Plot up to two data sets, one being a line plot and the other being 
    a histogram plot.
    
    pandas df: data1: The first data set to plot
    pandas df: data2: The second data set to plot
    string: title: The title of the plot
    """
    # Create Sub Plots
    figure, axs = plt.subplots(2, 2, figsize=(12,8))
    
    # Plot the first data
    axs[0,0].plot(data1, color = 'y', linestyle = 'dashed')
    axs[0,1].hist(data1, color = 'y', label='Happiness')
    
    # Plot the normal data 
    data1_norm_df = (data1 - data1.min()) / (data1.max() - data1.min())
    axs[1,0].plot(data1_norm_df, color='y', linestyle = 'dashed')
    axs[1,1].hist(data1_norm_df, color='y')
    
    # Plot the second data 
    try:
        # Plot the first data
        axs[0,0].plot(data2, color = 'b')
        axs[0,1].hist(data2, color='b', alpha=0.8, label='Comparing Data')

        # Plot the normal data 
        data2_norm_df = (data2 - data2.min()) / (data2.max() - data2.min())
        axs[1,0].plot(data2_norm_df, color='b')
        axs[1,1].hist(data2_norm_df, color='b', alpha=0.6)
            
    except Exception:
        # Do nothing
        pass
    
    leg = figure.legend(loc='center', bbox_to_anchor=(0.5, -0.10), shadow=False, ncol=2)
    
    # Create the font I want to use
    font1 = {'family':'serif','color':'black','size':20}
    font2 = {'family':'serif','color':'black','size':15}

    # Set the title and labels
    figure.suptitle(title, fontsize=18)
    axs[1,0].set_title("Normalized Line", fontsize=18)
    axs[1,1].set_title("Normalized Histogram", fontsize=18)
    figure.tight_layout();
```

## Regression Function
Below I have created a function that will create a Linear Regression model using the Happiness score vs the input factor. It will then predict multiple Happiness scores using a list of factor values I have specified. It will then return the R-Squared value from that model which I can use to help determine how good or bade the models are.

```python
def plot_reg(x, prediction_list):
    """ A function to plot a simple single linear regression.
    """
    X = df[[x]]
    y = df[["Score"]]

    reg = LinearRegression()
    model = reg.fit(X,y)
    
    # Display R-Squared value
    r_sqr = model.score(X,y)

    # Create the figure size
    plt.figure(figsize=(12,7))
    
    # Plot the regression plot
    sns.regplot(x=x,y="Score",data=df, scatter_kws = {'color':'r','s':12})
    
    # Display the values as we predict them
    print("A countries happiness score when their gdp value is:")
    for p in prediction_list:
        plt.vlines(x=p, ymin=2, ymax=model.predict([p]), color='orange')
        plt.text(p[0], 2, f'y={round(model.predict([p])[0][0], 4)}', fontsize = 10, color = 'b')
        print(f"{p[0]}:",model.predict([p])[0][0])

    # Plot title and show the graph
    plt.title(f"{x} Prediction Reg Plot")
    plt.show()
    
    return r_sqr;
```

## Multi Linear Regression
Now that we have seen the R-Squared values for the simple Linear regression. I want to see what combinations of factors can create the highest and most accurate model and R-Squared value. I will do this by starting with the highest R-Squared factor __GDP__ and then introduce one other factor at a time saving only the highest increase factor before continuing.

```python
def get_stats(formula, current_selected):
    # This will contain the best parameter (factor, value)
    best_factor = (None, 0)
    
    # Loop through all remaining allowed_factors
    for factor in ["Generosity", "Corruption", "Freedom", "Social", "Life_Expectancy", "GDP"]:
        # Skip if factor already in formula
        if factor in current_selected:
            continue
        
        # Build model and check if better
        model = smf.ols(formula=formula+factor, data=df).fit()
        if model.rsquared_adj > best_factor[1]:
            best_factor = factor, model.rsquared_adj
    
    return best_factor
```

```python
# k = 1
best_adf_r = [] # Best Addition Factor
best = get_stats('Score~', [])
best_adf_r.append(best)
print(best_adf_r)
X = df.drop("Score",axis=1)
y = df["Score"]
```

```python
# k = 2
best = get_stats('Score~GDP+', ['GDP'])
best_adf_r.append(best)
print(best_adf_r)
X = df.drop("Score",axis=1)
y = df["Score"]
```

```python
# k = 3
best = get_stats('Score~GDP+Freedom+', ['GDP', 'Freedom'])
best_adf_r.append(best)
print(best_adf_r)
X = df.drop("Score",axis=1)
y = df["Score"]
```

```python
# k = 4
best = get_stats('Score~GDP+Freedom+Social+', ['GDP', 'Freedom', 'Social'])
best_adf_r.append(best)
print(best_adf_r)
X = df.drop("Score",axis=1)
y = df["Score"]
```

```python
# k = 5
best = get_stats('Score~GDP+Freedom+Social+Life_Expectancy+', ['GDP', 'Freedom', 'Social', 'Life_Expectancy'])
best_adf_r.append(best)
print(best_adf_r)
X = df.drop("Score",axis=1)
y = df["Score"]
```

```python
# k = 6
best = get_stats('Score~GDP+Freedom+Social+Life_Expectancy+Generosity+', 
                 ['GDP', 'Freedom', 'Social', 'Life_Expectancy', 'Generosity'])
best_adf_r.append(best)
print(best_adf_r)
X = df.drop("Score",axis=1)
y = df["Score"]
```

### Plotting the R-Squared increse with each factor

```python
x_sqr = [r[0] if r[0] == "GDP" else f"+{r[0]}" for r in best_adf_r]
y_sqr = [r[1] for r in best_adf_r]
y_sqr = np.array(y_sqr)

plt.figure(figsize=(12,6))
plt.plot(x_sqr, y_sqr, color='r')
plt.title("Multi Factor Regression R-Squared Values")
plt.ylabel("r_squared_adj")
plt.xlabel("Factors");
```
